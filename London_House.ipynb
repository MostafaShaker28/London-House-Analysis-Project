##**Heat Map To Numeric Column**

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap', fontsize=14, fontweight='bold')
plt.show()
```

```python
df.describe()
```

##**Outlier values in apartment prices**

```python
Q1 = 1.161400e+06
Q3 = 2.390500e+06
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
```

```python
outliers_price = df[df['Price (£)'] > upper]
outliers_price
```

```python
fig = px.box(df, x='Price (£)', color_discrete_sequence=['red'],
             title='Outliers Detection in Property Prices')
fig.update_layout(title_font=dict(size=18, family='Arial', color='black'))
fig.show()
```

##**Histplot for the distribution of apartment prices**

```python
plt.figure(figsize=(10,6))
sns.histplot(
    data=df,
    x='Price (£)',
    bins=30,
    kde=True,
    color='#1f77b4',
    alpha=0.8
)

plt.title('Distribution of Property Prices in London', fontsize=16, fontweight='bold', color='#333')
plt.xlabel('Price (£)', fontweight = 'bold' ,fontsize=12)
plt.ylabel('Number of Properties', fontweight = 'bold' , fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

##**Histplot for the distribution of apartment sizes**

```python
plt.figure(figsize=(10,6))
sns.histplot(
    data=df,
    x='Square Meters',
    bins=30,
    kde=True,
    color='#1f77b4',
    alpha=0.8
)

plt.title('Distribution of Property Sizes in London', fontsize=16, fontweight='bold')
plt.xlabel('Square Meters (m²)', fontweight='bold', fontsize=12)
plt.ylabel('Number of Properties', fontweight='bold', fontsize=12)

plt.grid(alpha=0.3, color='gray')
plt.tight_layout()
plt.show()
```

##**Scatterplot of the relationship between area and price**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='Square Meters',
    y='Price (£)',
    color='#1f77b4',
    alpha=0.7,
    s=80
)

plt.title('Relationship Between Property Size and Price', fontsize=16, fontweight='bold')
plt.xlabel('Square Meters (m²)', fontweight='bold', fontsize=12)
plt.ylabel('Price (£)', fontweight='bold', fontsize=12)

plt.grid(alpha=0.3, color='gray')

plt.tight_layout()
plt.show()
```

##**lineplot of the average apartment price by building age**

```python
avg_buildig_age_byprice = df.groupby('Building Age')['Price (£)'].mean().sort_values(ascending=False)
avg_buildig_age_byprice
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(
    data=avg_buildig_age_byprice.reset_index(),
    x='Building Age',
    y='Price (£)',
    color='#1f77b4',
    linewidth=3
)

plt.title('Effect of Building Age on Property Price', fontsize=16, fontweight='bold')
plt.xlabel('Building Age (Years)', fontweight='bold', fontsize=12)
plt.ylabel('Average Price (£)', fontweight='bold', fontsize=12)

plt.grid(alpha=0.3, color='gray')

plt.tight_layout()
plt.show()
```

##**Pie chart to calculate the occupancy rate of apartments**

```python
building_status_counts = df['Building Status'].value_counts()
building_status_counts
```

```python
plt.figure(figsize=(8,8))

colors = ['#a6c8ff', '#1f77b4', '#0b3c5d']

plt.pie(
    building_status_counts,
    labels=building_status_counts.index,
    autopct='%1.1f%%',
    explode=[0.1,0,0],
    shadow=True,
    startangle=90,
    colors=colors,
    textprops={'color': 'black', 'fontsize': 12},
    wedgeprops={'edgecolor' : 'black', 'linewidth': 2}
)

plt.title('Building Status Distribution', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

##**Barplot of the effect of heating type on total price**

```python
heating_type_byprice = df.groupby('Heating Type')['Price (£)'].sum().sort_values(ascending=False)
heating_type_byprice
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x=heating_type_byprice.index,
    y=heating_type_byprice.values,
    hue=heating_type_byprice.index,
    palette='Blues_r'
)

plt.title('Total Property Prices by Property Type', fontsize=16, fontweight='bold')
plt.xlabel('Property Type', fontsize=12, fontweight='bold')
plt.ylabel('Total Price (£)', fontsize=12, fontweight='bold')
plt.xticks(fontweight='bold')

plt.tight_layout()
plt.show()
```
